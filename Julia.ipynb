Изобразите график функции, заданный параметрически:
$$
\begin{cases}
x(t) = 2a \cos ^2 t + b \cos t \\
y(t) = a \sin 2t + d \sin t , \quad t \in [0; 2 \pi]
\end{cases}
$$


```julia
Код на Julia
```

```julia
println("Введите параметр a:")
a = parse(Float64, readline())
println("Введите параметр b:")
b = parse(Float64, readline())

n_tochki = 300
x_list = Float64[]
y_list = Float64[]

for i in 0:n_tochki
    t = 2 * π * i / n_tochki
    x = 2 * a * cos(t)^2 + b * cos(t)
    y = a * sin(2*t) + b * sin(t)
    push!(x_list, x)
    push!(y_list, y)
end

x_str = join([string(round(x, digits=4)) for x in x_list], ", ")
y_str = join([string(round(y, digits=4)) for y in y_list], ", ")

open("param.asy", "w") do f
    println(f, "import graph;")
    println(f, "size(400);")
    println(f, "real[] x = {", x_str, "};")
    println(f, "real[] y = {", y_str, "};")
    
    println(f, "draw(graph(x, y), red + linewidth(1pt));")
    
    println(f, "xaxis(\"\$x\$\", Bottom, Ticks, Arrow);")
    println(f, "yaxis(\"\$y\$\", Left, Ticks, Arrow);")
    
    println(f, "label(\"\$a = $a, b = $b\$\", (max(x), max(y)), NE);")
end

run(`asy -f pdf param.asy`)    
run(`xdg-open param.pdf`)
```
