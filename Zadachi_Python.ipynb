Изобразите график функции, заданный параметрически:
$$
\begin{cases}
x(t) = 2a \cos ^2 t + b \cos t \\
y(t) = a \sin 2t + d \sin t , \quad t \in [0; 2 \pi]
\end{cases}
$$


Код на Python:

```python
import math
import subprocess # Данная библиотека используется для запуска кода на asymptote

# Задаём параметры кривой
a = float(input("Введите параметр а: "))
b = float(input("Введите параметр b: "))
n_tochki = 300 # Задаём количество точек для построения кривой
# Создаём массивы для координат х и у
x_list = []
y_list = []
# Вычисляем точки кривой
for i in range(n_tochki + 1):
    t = 2 * math.pi * i / n_tochki # Считаем значение параметра t в текущей точке
    x = 2 * a * math.cos(t)**2 + b * math.cos(t) # Вычисляем координаты х
    y = a * math.sin(2*t) + b * math.sin(t) # Вычисляем координаты у
    x_list.append(x)
    y_list.append(y)
# Подготавливаем данные для asymptote, преобразовывая списки чисел в строки, разделённые запятыми
x_str = ','.join([str(round(x,6)) for x in x_list])
y_str = ','.join([str(round(y,6)) for y in y_list])
# Объявляем многострочную f-строку, которая создаёт текст asymtote-кода
asy_code = f"""
import graph; // Библиотека для графиков
size(400); // Устнавливаем размер полотна графика

// Объявляем массивы действительных чисел, которые будут хранить координаты кривой
real[] x = {{{x_str}}};
real[] y = {{{y_str}}};

draw(graph(x, y), red); // функция graph(x, y) строит путь по массивам х и у, а функция draw рисует кривую по этому пути

// Рисуем оси
xaxis("$x$", axis=Bottom, ticks=Ticks, arrow=Arrow);
yaxis("$y$", axis=Left, ticks=Ticks, arrow=Arrow);

// Добавляем подписи с параметрами
label("$a = {a}, b = {b}$", (0,0), NE);
"""
# Сохраняем Asymptote-код
with open("param.asy", "w") as f: f.write(asy_code) 
# ЗАПУСК ASYMPTOTE
subprocess.run(["asy", "-f", "pdf", "param.asy"])
subprocess.run(["xdg-open", "param.pdf"])
```

Дан эллипс, заданный уравнением:
$$
\frac{x^2}{16} + \frac{y^2}{9} = 1
$$
Найдите координаты фокусов, уравнения директрис и эксцентриситет эллипса

```python
#!/usr/bin/env python3

"""Эллипс и его директрисы"""

import math
import subprocess # Данная библиотека используется для запуска кода на asymptote
import os

#Код на python

a = 4.0 # Параметры эллипса
b = 3.0
c = math.sqrt(a*a - b*b) # Вычисляем фокусное расстояние
e = c / a  # Находим значение эксцентриситета

print(f"a = {a}")
print(f"b = {b}")
print(f"Фокусное расстояние c = {c:.3f}")
print(f"Эксцентриситет e = {e:.3f}")

# Генирируем точки эллипса
n_tochki = 200
x_ellipse = []
y_ellipse = []

for i in range(n_tochki + 1):
    t = 2 * math.pi * i / n_tochki
    x = a * math.cos(t)
    y = b * math.sin(t)
    x_ellipse.append(x)
    y_ellipse.append(y)

F1 = (-c, 0) # Координаты левого фокуса эллипса
F2 = (c, 0) # Координаты правого фокуса эллипса

# Вычисляем директрисы
d_x = a*a / c # Расстояние от центра до директрисы
directrix1_x = -d_x # Координата левой директрисы
directrix2_x = d_x # Координата правой директрисы

n_check = 8 # вводим количество точек для проверки свойств эллипса
# Создаём пустые массивы для хранения координат точек эллписа
x_check = [] 
y_check = []
ratio1_vals = []
ratio2_vals = []

for i in range(n_check):
    idx = i * (n_tochki // n_check)
    x = x_ellipse[idx]
    y = y_ellipse[idx]
    
    # Вычисляем расстояние до фокусов
    MF1 = math.sqrt((x - F1[0])**2 + (y - F1[1])**2)
    MF2 = math.sqrt((x - F2[0])**2 + (y - F2[1])**2)
    
    # Вычисляем расстояние до директрис
    d1 = abs(x - directrix1_x)
    d2 = abs(x - directrix2_x)
    
    # Отношение расстояний
    ratio1 = MF1 / d1 
    ratio2 = MF2 / d2 
    
    x_check.append(x)
    y_check.append(y)
    ratio1_vals.append(ratio1)
    ratio2_vals.append(ratio2)
    
    print(f"Точка ({x:.2f}, {y:.2f}): MF/d = {ratio1:.4f}, {ratio2:.4f}")

#ASYMPTOTE-КОД
# Подготавливаем данные для asymptote, преобразовывая списки чисел в строки, разделённые запятыми
x_str = ', '.join([f'{xi:.6f}' for xi in x_ellipse])
y_str = ', '.join([f'{yi:.6f}' for yi in y_ellipse])
x_check_str = ', '.join([f'{xi:.6f}' for xi in x_check])
y_check_str = ', '.join([f'{yi:.6f}' for yi in y_check])
ratio1_str = ', '.join([f'{r:.4f}' for r in ratio1_vals])
ratio2_str = ', '.join([f'{r:.4f}' for r in ratio2_vals])

# Объявляем многострочную f-строку, которая создаёт текст asymtote-кода
asy_code = f"""
import graph; // Библиотека для графиков
size(500); // Устнавливаем размер полотна графика

// Объявляем массивы действительных чисел, которые будут хранить координаты точек эллипса
real[] x = {{{x_str}}};
real[] y = {{{y_str}}};

// Передаём координаты фокусов эллипса
pair F1 = ({F1[0]:.6f}, {F1[1]:.6f});
pair F2 = ({F2[0]:.6f}, {F2[1]:.6f});

// Передаём координаты директрисы эллипса
real d1_x = {directrix1_x:.6f};
real d2_x = {directrix2_x:.6f};

int n_check = {n_check};
real[] x_check = {{{x_check_str}}};
real[] y_check = {{{y_check_str}}};
real[] ratio1 = {{{ratio1_str}}};
real[] ratio2 = {{{ratio2_str}}};

// Рисуем эллипс
draw(graph(x, y), blue, 
     L=Label("Ellipse", position=Relative(0.3), align=NW)); // функция graph(x, y) строит путь по массивам х и у, а функция draw рисует эллпис по этому пути

// Рисуем оси 
xaxis("$x$", BottomTop, LeftTicks(Step=2));
yaxis("$y$", LeftRight, RightTicks(Step=2));

// Отображаем фокусы
dot(F1, red); // отмечаем координату левого фокуса на графике
dot(F2, red); // отмечаем координату правого фокуса на графике
label("$F_1$", F1 + (-0.3, 0.3), red); // добавляем подписи на график
label("$F_2$", F2 + (0.3, 0.3), red);

// Отображаем директрисы
// рисуем пунктирную линию от точки (d1_x, -4) до (d1_x, 4)
draw((d1_x, -4) -- (d1_x, 4), heavygreen + dashed, L=Label("directrix", position=Relative(0.5), align=N)); 
// рисуем пунктирную линию от точки (d2_x, -4) до (d2_x, 4)
draw((d2_x, -4) -- (d2_x, 4), heavygreen + dashed, L=Label("directrix", position=Relative(0.5), align=N)); 

for (int i = 0; i < n_check; ++i) {{
    real x0 = x_check[i];
    real y0 = y_check[i];
    real r1 = ratio1[i];
    real r2 = ratio2[i];
    
    dot((x0, y0), black); // рисуем точку по координате (х0; у0)
    
    // С помощью функции draw соединыем точки (х0; у0) и F1, (х0; у0) и F2
    draw((x0, y0) -- F1, gray + dotted); // dotted указывает, что линия пунктирная
    draw((x0, y0) -- F2, gray + dotted);
    
    // Добавляем подписи с формулами
    if (i % 2 == 0) {{
        label(scale(0.6)*format("$\\frac{{MF_1}}{{d_1}} = %.3f$", r1), (x0 + 0.2, y0 + 0.5), black);
    }} else {{
        label(scale(0.6)*format("$\\frac{{MF_2}}{{d_2}} = %.3f$", r2), (x0 - 0.2, y0 - 0.5), black);
    }}
}}

// Добавляем подписи с параметрами эллипса
label("$a = 4$, $b = 3$", (-3, 5), black);
label("$c = \\sqrt{{a^2-b^2}} = {c:.3f}$", (-3, 4.5), black);
label("$\\epsilon = c/a = {e:.3f}$", (-3, 4), black);

shipout(bbox(2mm));
"""

# Сохраняем Asymptote-код
with open("ellipse.asy", "w") as f: f.write(asy_code)

# ЗАПУСК ASYMPTOTE
result = subprocess.run(["asy", "-f", "pdf", "ellipse.asy"], capture_output=True, text=True)
subprocess.run(["xdg-open", "ellipse.pdf"])
```
