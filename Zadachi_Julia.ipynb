Изобразите график функции, заданный параметрически:
$$
\begin{cases}
x(t) = 2a \cos ^2 t + b \cos t \\
y(t) = a \sin 2t + d \sin t , \quad t \in [0; 2 \pi]
\end{cases}
$$


```julia
Код на Julia
```

```julia
# Задаём параметры кривой
println("Введите параметр a:")
a = parse(Float64, readline())          # Считываем строку и преобразуем в число с плавающей точкой
println("Введите параметр b:")
b = parse(Float64, readline())
n_tochki = 300                          # Задаём количество точек для построения кривой
                                        # Создаём массивы для координат х и у
x_list = Float64[]                       
y_list = Float64[]
                                        # Вычисляем точки кривой
for i in 0:n_tochki
    t = 2 * π * i / n_tochki            # Считаем значение параметра t в текущей точке
    x = 2 * a * cos(t)^2 + b * cos(t)    # Вычисляем координаты х
    y = a * sin(2t) + b * sin(t)          # Вычисляем координаты y
    push!(x_list, x)                      # Добавляем x в массив
    push!(y_list, y)                      # Добавляем y в массив
end

# Подготавливаем данные для asymptote, преобразовывая списки чисел в строки, разделённые запятыми
x_str = join([string(x) for x in x_list], ',')
y_str = join([string(y) for y in y_list], ',')

# Объявляем многострочную f-строку, которая создаёт текст asymtote-кода
asy_code = """
import graph;                  // Библиотека для графиков
size(400);                    // Устнавливаем размер полотна графика

// Объявляем массивы действительных чисел, которые будут хранить координаты кривой
real[] x = {$x_str};          
real[] y = {$y_str};

// Функция graph(x, y) строит путь по массивам х и у, а функция draw рисует кривую по этому пути
draw(graph(x, y), red);

// Рисуем оси
xaxis("\$x\$", axis=Bottom, ticks=Ticks, arrow=Arrow);
yaxis("\$y\$", axis=Left, ticks=Ticks, arrow=Arrow);

// Добавляем подписи с параметрами
label("\$a = $a, b = $b\$", (0,0), NE);  
"""

# Запись Asymptote кода в файл
open("param.asy", "w") do f
    write(f, asy_code)       
end
# Запуск Asymptote для создания PDF и открытие полученного файла
run(`asy -f pdf param.asy`)     
run(`xdg-open param.pdf`)       

# подключение пакетов для правильного отображения строк текста asymtote-кода
using Markdown
display(Markdown.Code("asy", asy_code))
```

Дан эллипс, заданный уравнением:
$$
\frac{x^2}{16} + \frac{y^2}{9} = 1
$$
Найдите координаты фокусов, уравнения директрис и эксцентриситет эллипса

```julia
Код на Julia
```

```julia
# Задание параметров эллипса
a = 4.0
b = 3.0
# Вычисление фокусного расстояния и эксцентриситета 
c = sqrt(a * a - b * b)
e = c / a

println("a = ", a)
println("b = ", b)
println("Фокусное расстояние c = ", round(c; digits=3))
println("Эксцентриситет e = ", round(e; digits=3))

# Генерация точек эллипса с параметрическим представлением
n_tochki = 200               # количество точек на эллипсе
x_ellipse = Float64[]        # массивы для координат X и Y
y_ellipse = Float64[]
for i in 0:n_tochki
    t = 2π * i / n_tochki    
    push!(x_ellipse, a * cos(t))
    push!(y_ellipse, b * sin(t))
end

# Координаты фокусов эллипса
F1 = (-c, 0.0)
F2 = (c, 0.0)   

# Вычисление координат директрис
d_x = a * a / c                
directrix1_x = -d_x      
directrix2_x = d_x  

# Вводим количество точек для проверки свойств эллипса
n_check = 8        
x_check = Float64[]   
y_check = Float64[]
ratio1_vals = Float64[]         
ratio2_vals = Float64[]        

for i in 0:n_check-1
    idx = i * (n_tochki ÷ n_check) + 1  
    x = x_ellipse[idx]
    y = y_ellipse[idx]

    # Расстояния до фокусов
    MF1 = sqrt((x - F1[1])^2 + (y - F1[2])^2)
    MF2 = sqrt((x - F2[1])^2 + (y - F2[2])^2)

    # Расстояния до соответствующих директрис
    d1 = abs(x - directrix1_x)
    d2 = abs(x - directrix2_x)

    # Вычисление отношений
    ratio1 = MF1 / d1
    ratio2 = MF2 / d2

    # Сохраняем данные
    push!(x_check, x)
    push!(y_check, y)
    push!(ratio1_vals, ratio1)
    push!(ratio2_vals, ratio2)

    println("Точка (", round(x; digits=2), ", ", round(y; digits=2),"): MF/d = ", round(ratio1; digits=4), ", ", round(ratio2; digits=4))
end

# Подготовка данных для вставки в Asymptote 
x_str = join(string.(x_ellipse), ", ")
y_str = join(string.(y_ellipse), ", ")
x_check_str = join(string.(x_check), ", ")
y_check_str = join(string.(y_check), ", ")
ratio1_str = join(string.(ratio1_vals), ", ")
ratio2_str = join(string.(ratio2_vals), ", ")

# Формирование кода для Asymptote
asy_code = """
import graph;
size(500);

real[] x = {$x_str};
real[] y = {$y_str};

pair F1 = ($(F1[1]), $(F1[2]));
pair F2 = ($(F2[1]), $(F2[2]));

real d1_x = $directrix1_x;
real d2_x = $directrix2_x;

int n_check = $n_check;
real[] x_check = {$x_check_str};
real[] y_check = {$y_check_str};
real[] ratio1 = {$ratio1_str};
real[] ratio2 = {$ratio2_str};

draw(graph(x, y), blue, L=Label("Ellipse", position=Relative(0.3), align=NW));

xaxis(Label("\$x\$"), Bottom, ticks=LeftTicks(Step=2));
yaxis(Label("\$y\$"), Left, ticks=RightTicks(Step=2));

dot(F1, red);
dot(F2, red);
label("\$F_1\$", F1 + (-0.3, 0.3), red);
label("\$F_2\$", F2 + (0.3, 0.3), red);

draw((d1_x, -4) -- (d1_x, 4), heavygreen + dashed, L=Label("directrix", position=Relative(0.5), align=N));
draw((d2_x, -4) -- (d2_x, 4), heavygreen + dashed, L=Label("directrix", position=Relative(0.5), align=N));

for (int i = 0; i < n_check; ++i) {
    real x0 = x_check[i];
    real y0 = y_check[i];
    real r1 = ratio1[i];
    real r2 = ratio2[i];
    
    dot((x0, y0), black);
    
    draw((x0, y0) -- F1, gray + dotted);
    draw((x0, y0) -- F2, gray + dotted);
    
    if (i % 2 == 0) {
        label(scale(0.6)*format("\$\\frac{MF_1}{d_1} = %.3f\$", r1), (x0 + 0.2, y0 + 0.5), black);
    } else {
        label(scale(0.6)*format("\$\\frac{MF_2}{d_2} = %.3f\$", r2), (x0 - 0.2, y0 - 0.5), black);
    }
}

label("\$a = 4,\\; b = 3\$", (-3, 5), black);
label("\$c = \\sqrt{a^2-b^2} = $(round(c, digits=3))\$", (-3, 4.5), black);
label("\$\\epsilon = c/a = $(round(e, digits=3))\$", (-3, 4), black);

shipout(bbox(2mm));
"""

# Запись Asymptote‑кода в файл
write("ellipse.asy", asy_code)

# Запуск Asymptote для создания PDF и открытие полученного файла
run(`asy -f pdf ellipse.asy`)
run(`xdg-open ellipse.pdf`)

# подключение пакетов для правильного отображения строк текста asymtote-кода
using Markdown
display(Markdown.Code("asy", asy_code))
```
